```python
def removing_outliers_from_aqi(text):
    q1=df['aqi_value'].quantile(0.25)
    q3=df['aqi_value'].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    if text > upper:
        return upper
    elif text < lower:
        return lower
    else:
        return text
```

```python
df['aqi_value']=df['aqi_value'].apply(removing_outliers_from_aqi)
```

```python
def extract_features(text):
    text=text.split('-')
    if int(text[2]) >=int(text[2]):
        if (int(text[2])==2024 and int(text[1])==12) or (int(text[2])==2025):
            return '-'.join(text)
    else:
        return None
```

```python
southern_india_states = [
    "Andhra Pradesh",
    "Karnataka",
    "Kerala",
    "Tamil Nadu",
    "Telangana",
    "Puducherry"
]
```

```python
south_states=df[df['state'].isin(southern_india_states)]
```

```python
south_states.head()
```

```python
south_states['year']=pd.to_datetime(south_states['date']).dt.year
```

```python
south_states['year'].head()
```

```python
def get_pollutant_rating(pollutant: str) -> int:
    """
    Returns a rating from 1 (least prominent) to n (most prominent),
    based on general AQI impact and concentration levels.
    """
    pollutant_rank = {
        "NH3": 1,
        "SO2": 2,
        "O3": 3,
        "CO": 4,
        "NO2": 5,
        "PM2.5": 6,
        "PM10": 7  # Highest impact, especially in Indian cities
    }
    
    return pollutant_rank.get(pollutant.upper(), 0)  # 0 = unknown pollutant
```

```python
south_states['Prominent Rating']=south_states['prominent_pollutants'].apply(get_pollutant_rating)
```

```python
south_states.head()
```

```python
# Function to get top 2 and bottom 2 prominent pollutants per state
def get_top_bottom_pollutants(df):
    results = []

    for state, group in df.groupby('state'):
        # Sort pollutants by Prominent Rating (descending for top, ascending for bottom)
        top2 = group.sort_values(by='Prominent Rating', ascending=False).head(2)
        bottom2 = group.sort_values(by='Prominent Rating', ascending=True).head(2)

        # Add to result
        results.append({
            'state': state,
            'top_pollutants': top2['prominent_pollutants'].tolist(),
            'bottom_pollutants': bottom2['prominent_pollutants'].tolist()
        })

    return pd.DataFrame(results)

# Apply function to filtered data
result_states = get_top_bottom_pollutants(south_states)
```
